## Face Identificationusing VGGFace (resnet50)

```python
def detect_face(filename):
    # create detector
    detector = MTCNN()
    
    # read image
    pixels = plt.imread(filename)
    
    # detect face in the input image
    results = detector.detect_faces(pixels)
    x1, y1, width, height = results[0]['box']
    x2, y2 = x1 + width, y1 + height

    # get face pixels from input image
    face_pixels = pixels[y1:y2, x1:x2]
    
    # convert pixels array to image and resize to (224, 224)
    face_image = Image.fromarray(face_pixels)
    face_image = face_image.resize([224,224])
    
    # again back to array
    face_array = asarray(face_image)
    
    return face_array

ralph_klein = detect_face('ralph_klein.jpg')
plt.imshow(ralph_klein)
plt.show()
```

```python
model = VGGFace(model='resnet50')
print('Input: %s' % model.inputs)
print('Output: %s' % model.outputs)
```

```python
def predict_image(input_image):
    sample_image = input_image.astype('float32')
    X_test = expand_dims(sample_image, axis=0)
    X_test = preprocess_input(X_test, version=2)

    yhat = model.predict(X_test)
    results = decode_predictions(yhat)

    return results

predict_image(ralph_klein)
```

## Face Verification using VGGFace (resnet50)

```python
model = VGGFace(model='resnet50', include_top=False, input_shape=(224, 224, 3), pooling='avg')
```

```python
def get_face_features(filenames):
    faces = [detect_face(f) for f in filenames]

    samples = asarray(faces, 'float32')
    samples = preprocess_input(samples, version=2)
    
    yhat = model.predict(samples)
    
    return faces, yhat

faces, embeddings = get_face_features(['shahram_akbarinasaji.jpg', 'shahram_akbarinasaji2.jpg', 'shahram_akbarinasaji3.jpg',
                                       'channing_tatum.jpg', 'ralph_klein.jpg'])
```

```python
def is_match(known_embedding, candidate_embedding, thresh=0.5):
    # calculate distance between embeddings
    score = cosine(known_embedding, candidate_embedding)
    
    if score <= thresh:
        print('> face is a Match (%.3f <= %.3f)' % (score, thresh))
    else:
        print('> face is NOT a Match (%.3f > %.3f)' % (score, thresh))
```

```python
def compare_embeddings(known_index, candidate_index, faces, embeddings, threshold):
    plt.subplot(1,2,1)
    plt.imshow(faces[known_index])
    
    plt.subplot(1,2,2)
    plt.imshow(faces[candidate_index])

    is_match(embeddings[known_index], embeddings[candidate_index], threshold)
```

```python
compare_embeddings(0, 1, faces, embeddings, 0.5)
```

```python
compare_embeddings(0, 2, faces, embeddings, 0.5)
```

```python
compare_embeddings(0, 3, faces, embeddings, 0.5)
```
