```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from dataprep.eda import plot
```

```python
df = pd.read_csv('results_t0.20.csv')

plot(df)
```

```python
not_same = len(df[df['is_same'] == 0])
not_same_predicted = len(df[df['predicted'] == 0])
same = len(df[df['is_same'] == 1])
same_predicted = len(df[df['predicted'] == 1])

arr = np.array([not_same, not_same_predicted, same, same_predicted])
arr_types = np.array(['Different Face', 'Different Face', 'Same Face', 'Same Face'])
arr_ver_count = np.array(['# Real', '# Predicted', '# Real', '# Predicted'])

df_temp = pd.DataFrame({'Face Pair': arr_types, 'Count': arr, 'Type': arr_ver_count})

sns.barplot(x='Face Pair', y='Count', data=df_temp, hue='Type')

plt.title('Face Verification (t = 0.20)', fontsize=15)
plt.legend(loc='lower right')


print('Accuracy: %2.5f' % accuracy_score(df['is_same'], df['predicted']))
print('Precision: %2.5f' % precision_score(df['is_same'], df['predicted']))
print('Recall: %2.5f' % recall_score(df['is_same'], df['predicted']))
print('F1-Score: %2.5f'  % f1_score(df['is_same'], df['predicted']))
```
